System of equations representation in matrix

```python
import numpy as np

a = np.array([[-1,3],[3,2]], dtype = np.dtype(float))

b = np.array([7,1],dtype = np.dtype(float))

print("Matrix A")
print(a)
print("Matrix B")
print(b)

size_A = a.size
size_B = b.size
shape_A = a.shape
shape_B = b.shape

print(size_A)
print(size_B)
print(shape_A)
print(shape_B)
```

Solving Linear Equations using Numpy methods

```python
import numpy as np

a = np.array([[-1,3],[3,2]], dtype = np.dtype(float))

b = np.array([7,1],dtype = np.dtype(float))

x = np.linalg.solve(a,b)
print(x)

# Determinant 

y = np.linalg.det(a)
print(y)

```
