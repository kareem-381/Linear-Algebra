Another way of implementing Array using arange method


```python
# Create an array with 3 integers 
import numpy as np
b = np.arange(3)
print(b)


# We can create an array using arange function in which three paramters are passed
# 1. The starting value then end value and last para is step how much step you want to add 

c = np.arange(1,20,3)
print(c)

# We have an another method called linspace it take same parameter but its job is different
# if we want to divide an array into x evenly spaces then we use this method
# default is float values np.float64

d = np.linspace(0,100,5)
print(d)

# we can change its datatype to int we have a attribute dtype = ?

e = np.linspace(0,100,5,dtype = int)
print(e)
```

Built in Methods

```python
import numpy as np

# ones, zeros, empty and random
# these 4 methods always required one paramter

a = np.ones(3,dtype= int)
b = np.zeros(3, dtype = int)
c = np.empty(3, dtype= int)
d = np.random.rand(3)

print(a)
print(b)
print(c)
print(d)
```

Multi Dimensional Array

```python
# There are different ways to create multi dimensional array
# first is simple one

a = np.array([[3,2,1],[4,5,6]])
print(a)

# second one is reshaping
b = np.array([1,2,3,4,5,6])

multi_Dimension= np.reshape(b,
                            (2,3)
                            )
print(multi_Dimension)

# we can store the size, shape and dimension

size = multi_Dimension.size
shape = multi_Dimension.shape
dimension = multi_Dimension.ndim

print("The size of the array is ",size)
print("The shape of the array is ",shape)
print("The Dimension of the array is ",dimension)
```

Array Math Operation

```python
# There are multiple operations we can perform

import numpy as np


arr1= np.array([1,2,3])
arr2 = np.array([3,4,6])

addition = arr1 + arr2
subtraction = arr1 - arr2
multiplication = arr1 * arr2

print("Addition")
print(addition)
print("Subtraction")
print(subtraction)
print("Multiplication")
print(multiplication)

# we can multiply a scalar with a vector


vector = np.array([1,2])
vector = vector * 1.6
print(vector)
```

Slicing

```python
# Slicing gives you a sublist of elements that you specify from the array. The slice notation 
# specifies a start and end value, and copies the list from start up to but not including the end (end-exclusive).

# The syntax is:

# array[start:end:step]

# If no value is passed to start, it is assumed start = 0, if no value is passed for end, 
# it is assumed that end = length of array and if no value is passed to step, it is assumed step = 1.


a = np.array([2,5,8,1,6,9])

slice_array= a[:4]
print(slice_array)
```

Stacking

```python
import numpy as np

a1 = np.array([[1,2],
               [3,4]])

a2 = np.array([[5,6],[8,9]])

# vertical takes only one argument
vertical_Stack= np.vstack((a1,a2))
print(vertical_Stack)

# horizontal takes only one argument

horizontal_stack = np.hstack((a1,a2))
print(horizontal_stack)
```
